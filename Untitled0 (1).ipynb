```python
import pandas as pd
import pycaret
from pycaret.classification import *
from ucimlrepo import fetch_ucirepo
```

```python
# Fetch dataset
wine_quality = fetch_ucirepo(id=186)

# Data (as pandas dataframes)
X = wine_quality.data.features
y = wine_quality.data.targets

# Combine features and target for PyCaret
data = pd.concat([X, y], axis=1)
```

```python
s = setup(data, target='quality', session_id=123)
```

```python
best = compare_models()
```

```python
plot_model(best, plot='confusion_matrix')
plot_model(best, plot='auc')
plot_model(best, plot='feature')
```

```python
evaluate_model(best)
```

```python
holdout_pred = predict_model(best)
holdout_pred.head()
```

```python
save_model(best, 'my_first_pipeline')
loaded_best_pipeline = load_model('my_first_pipeline')
```

```python
#Tune a Decision Tree model
dt = create_model('dt')
tuned_dt = tune_model(dt, optimize='F1')
```

```python
bagged_dt = ensemble_model(dt, method='Bagging')
boosted_dt = ensemble_model(dt, method='Boosting')
```

```python
stacked_models = stack_models([best, dt, bagged_dt])
blended_model = blend_models([best, dt, bagged_dt])
```

```python
final_model = finalize_model(best)
```
